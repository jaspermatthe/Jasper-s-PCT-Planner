```python
import geopandas as gpd
import folium
from shapely.geometry import LineString, MultiLineString, Point
import matplotlib.pyplot as plt
import numpy as np
```

```python
pct_trail_file = "PCT_Trail_Points.geojson"

points_gdf = gpd.read_file(pct_trail_file)
```

```python
import folium

# Create a map centered around a specific location (e.g., the first point in points_gdf)
m = folium.Map(location=[points_gdf.iloc[0]['geometry'].y, points_gdf.iloc[0]['geometry'].x], zoom_start=10)

# Iterate over each row in points_gdf
for index, row in points_gdf.iterrows():
    
    # Check if the point is a campsite
    if row['Campsite']:
        popup_text = f"Title: {row['OWNERNAME']}<br>SOBO Mi: {row['SOBO Mi']}<br>"
        # Add a marker for the campsite
        folium.Marker(location=[row['geometry'].y, row['geometry'].x], popup=popup_text, tooltip=row['OWNERNAME'], icon=folium.Icon(color='red', icon='tent', prefix='fa')).add_to(m)
    if row['Ressuply']:
        popup_text = f"Title: {row['Ressuply']}<br>SOBO Mi: {row['SOBO Mi']}<br>"
        folium.Marker(location=[row['geometry'].y, row['geometry'].x], popup=popup_text, tooltip=row['Ressuply'], icon=folium.Icon(color='green', icon='box', prefix='fa')).add_to(m)
    if row['Turning/Shuttle']:
        popup_text = f"Title: {row['Turning/Shuttle']}<br>SOBO Mi: {row['SOBO Mi']}<br>"
        folium.Marker(location=[row['geometry'].y, row['geometry'].x], popup=popup_text, tooltip=row['Turning/Shuttle'], icon=folium.Icon(color='pink')).add_to(m)

# Display the map
m
```
